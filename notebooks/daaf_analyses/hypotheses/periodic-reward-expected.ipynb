```python
import os
import os.path
import json

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
k = 10000
x = np.arange(0, k)
y = np.log(k + 2) ** 2 - np.log(x + 2) ** 2
z = np.log(k + 2) ** 2 - np.log(x + 2) ** 1.8
data = pd.DataFrame({"x": x, "y": y, "z": z})
```

```python
fig, ax = plt.subplots(1)
sns.lineplot(data=data, x="x", y="y", ax=ax)
sns.lineplot(data=data, x="x", y="z", ax=ax)
ax.set_title("RMSE by Episode")
ax.set_xlabel("Episode")
ax.set_ylabel("RMSE")
ax.legend(["Regular Reward", "Periodic Reward"])
```

```python
fig.savefig("../periodic-reward", dpi=120, transparent=True)
```
